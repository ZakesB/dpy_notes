## Question 4

```python
#Complete the code

print(data_df.columns)
labels_df = data_df['class']
features_df = data_df[['sepal_length_cm', 'sepal_width_cm',
                             'petal_length_cm', 'petal_width_cm']]
```

## Question 5

```python
#Complete the code where there is (...)

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test =train_test_split(features_df,labels_df,random_state=1,test_size=0.3)
```

## Question 6

```python
#Complete the code where there is (...)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=90)
knn.fit(x_train,y_train)
predictions = knn.predict(x_test)
score = accuracy_score(y_test,predictions)
score
```

## Question 7

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree_model = DecisionTreeClassifier(random_state=0)
tree_model.fit(x_train, y_train)
predictions = tree_model.predict(x_test)
score = accuracy_score(y_test, predictions)
print(score)
```
